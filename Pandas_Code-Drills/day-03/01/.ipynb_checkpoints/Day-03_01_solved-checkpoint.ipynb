```python
# import the module pandas
import pandas as pd
```

```python
# Set the path of the BAC stock data file to be loaded from the Data folder.
    # Define the path name as GSPC_path
BAC_path = "Data/BAC.csv"


# Define a variable "BAC_df" and store the BAC.csv into Pandas DataFrame using the BAC_path
BAC_df = pd.read_csv(BAC_path, encoding="utf-8")
BAC_df.head()
```

```python
# Set the path of the GSPC stock data file to be loaded from the Data folder. 
    # Define the path name as GSPC_path
    # Side note: ^GSPC is the ticket name for the S&P500.
GSPC_path = "Data/GSPC.csv"


# Define a variable "GSPC_df" and store the GSPC.csv into Pandas DataFrame using the GSPC_path
GSPC_df = pd.read_csv(GSPC_path, encoding="utf-8")
GSPC_df.head()
```

```python
# Set a variable "merge_table" and merge the dataframes BAC_df and GSPC_df using an inner join.
    # Since the two dataframes have the same column names, use the suffixes option for merge.
    # http://pandas.pydata.org/pandas-docs/stable/generated/pandas.merge.html#pandas.merge
    # Example: merge(df1, df2, on="something", suffixes=('_BAC', '_GSPC')
merge_table = pd.merge(BAC_df, GSPC_df, on="Date", suffixes=('_BAC', '_GSPC'))
merge_table.head()
```

```python
# Create a new column in merge_table called "Price Change_BAC" 
    # and set it equal to column Close_BAC - Open_BAC
merge_table["Price Change_BAC"]= merge_table["Close_BAC"]- merge_table["Open_BAC"]

# Create a new column in merge_table called "Price Change_GSPC" 
    # and set it equal to column Close_GSPC - Open_GSPC
merge_table["Price Change_GSPC"]= merge_table["Close_GSPC"]- merge_table["Open_GSPC"]

# Show the first 10 items using Pandas head function.
merge_table.head(10)
```

```python
# Define a new dataframe called organized_df and list only the Open, Close, Net columns
organized_df = merge_table[["Date", "Price Change_BAC", "Price Change_GSPC"]]

# Show the first 10 items using Pandas head function.
organized_df.head(10)
```

```python
# Use the corr function on organized_df to find the correlation of the two dataframes.
    # By default a Pearson correlation coefficient will be applied.
    # https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.corr.html
    # https://en.wikipedia.org/wiki/Pearson_correlation_coefficient

organized_df.corr()
```
