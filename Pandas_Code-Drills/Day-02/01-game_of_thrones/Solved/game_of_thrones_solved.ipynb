```python
# Import Dependencies
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
# Save file path to variable
got = "../Resources/game_of_thrones.csv"
```

```python
# Read csv with Pandas
got_pd = pd.read_csv(got)
got_pd.head()
```

```python
# Collect a list of all the unique values in "Allegiances"
got_pd["Allegiances"].unique()
```

```python
# Create a new table with only the Allegiances and Names
got_houses_df = got_pd[["Allegiances", "Name"]]
got_houses_df.head()
```

```python
# Create a GroupBy object based on the "Allegiances" column in the newley created table
got_houses = got_houses_df.groupby(["Allegiances"])

# Count the amount of deaths that has occured in each house
# Rename the "Name" column to "Death Count"
house_deaths = got_houses.count()
house_deaths.rename(columns={"Name" : "Death Count"}, inplace=True)
house_deaths
```

```python
# Sort by most deaths and reset index
most_deaths = house_deaths.sort_values(["Death Count"], ascending=False)
most_deaths.reset_index(inplace=True)
most_deaths
```

```python
# Save the death count for the Night's Watch, using `iloc` to filter
nights_watch = most_deaths.iloc[1]
nights_watch
```
