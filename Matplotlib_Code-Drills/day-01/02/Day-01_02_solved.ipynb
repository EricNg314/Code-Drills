```python
# Import numpy, matplotlib, and pandas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the csv from Resources folder as school_ranking_df
school_ranking_df = pd.read_csv("Resources/cwurData.csv")

# Show the first 5 rows of school_ranking_df
school_ranking_df.head(5)
```

```python
# Declare a variable called institution_avg_score_df 
    # Set it with the group of institution, average of score
    # Hint: When grouping, use as_index=False to retain column names.
institution_avg_score_df = school_ranking_df.groupby("institution", as_index=False)["score"].mean().sort_values("score", ascending=False)

# Show the first 10 rows of school_ranking_df
institution_avg_score_df.head(10)
```

```python
# Declare a variable called top_avg_sorted_df 
    # sort institution_avg_score_df by score from Highest to Lowest.
    # then save only the top 10 rows
top_avg_sorted_df = institution_avg_score_df.sort_values("score", ascending=False).head(10)

# Show the rows of top_avg_sorted_df
top_avg_sorted_df
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(top_avg_sorted_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
# Bars should indicate score
# Horizontal axis should show the institutions
plt.figure(figsize=(10,8))
plt.bar(x_axis, top_avg_sorted_df["score"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, top_avg_sorted_df["institution"], rotation="vertical")
```

```python
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(70, 100)
```

```python
# Set a title and labels
plt.title("Average Score of top 10 Institution 2012-2015")
plt.xlabel("Institution")
plt.ylabel("Average Score")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("./images/avg_top10_insitutions.png")
plt.show()
```
