```python
# Import Numpy for calculations and matplotlib for charting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Below is the average price of avocado in a week for the year of 2017.
region = ["Las Vegas","Los Angeles","Sacramento","San Diego","San Francisco"]
average_2017 = [1.4178,1.3773,1.7740,1.5845,1.9659]

x_axis = np.arange(len(average_2017))
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis, average_2017, color="#33C1FF", align="center")
```

```python
# Create the ticks for your bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, region)
```

```python
# Set the limits of the y axis between 0.5 to the max of the average with an additional 0.5
plt.ylim(0.5, max(average_2017)+0.5)
```

```python
# Give the chart a title, x label, and y label
plt.title("Average Avocado Price 2017")
plt.xlabel("Regions")
plt.ylabel("Average Price")
```
